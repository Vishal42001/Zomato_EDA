# Zomato DATA Analysis 

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

# Import the dataset

```python
dataframe = pd.read_csv('C:/Users/kvish/OneDrive/Desktop/Zomato_project/Zomato.csv')
print (dataframe)
```

```python
dataframe
```

```python

def handleRate (value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Type Of Restaurant

```python

sb.countplot(x=dataframe['listed_in(type)']) 
plt.xlabel("Type of Restaurant")
```

```python

grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="purple", marker="o")
plt.xlabel("Type Of Restaurant", c="blue", size=15)
plt.ylabel("Votes", c="blue", size=15)
```

```python

plt.hist(dataframe ['rate'],bins=8) 
plt.title("Ratings Distribution")
plt.show()
```

```python

couple_data=dataframe [ 'approx_cost(for two people)']
sb.countplot(x=couple_data)
sb.countplot(x = couple_data, palette = "Set1")
```

```python

plt.figure(figsize = (6,6))
sb.boxplot(x = 'online_order', y = 'rate', data=dataframe)
```
